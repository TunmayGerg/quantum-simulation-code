```python
'''
This code is simply instantiating and using a couple of basic tools in QuTiP, if it runs without error you should be all set.
'''

zero_state = qt.basis(2,0)
plus_state = (qt.basis(2,0) + qt.basis(2,1)).unit()

print(f"<0|+> = {zero_state.overlap(plus_state)}")
print(f"|<0|+>|^2 = {np.abs(zero_state.overlap(plus_state))**2}")

print(f"<0|X|0> = {qt.expect(qt.sigmax(),zero_state)}")
print(f"<+|X|+> = {qt.expect(qt.sigmax(),plus_state)}")

print("\nX acting on |0>:")
print(f"X|0> = {qt.sigmax()*zero_state}")

oscillator_trunc = 5
a = qt.destroy(oscillator_trunc)
a_dag = a.dag()

print("\nLet's try to compute [a,adag]:")
commutator = a* a_dag - a_dag * a
print(commutator)
print("This should be the identity operator, but we see that with finite truncation of our Hilbert space, we can get get artifacts.")


print("\nAll done!")
```
