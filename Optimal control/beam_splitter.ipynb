```python
import numpy as np
import matplotlib.pyplot as plt
import qutip as qt
from helpful_functions import *
```

```python
n1 = 3 # number of Fock states for mode 1
n2 = 3 # number of Fock states for mode 2

a1 = qt.tensor(qt.destroy(n1), qt.qeye(n2)) # annihilation operator for mode 1
a2 = qt.tensor(qt.qeye(n1), qt.destroy(n2)) # annihilation operator for mode 2

g = 1 * np.pi # beamsplitter strength
detuning = 0 * 2 * np.pi # detuning of the beamsplitter
phase = 0 # phase of the beamsplitter in radians

H = (g/2) * (a1.dag() * a2 * np.exp(1j * phase) + a1 * a2.dag() * np.exp(-1j * phase)) + detuning * a2.dag() * a2
U = (-1j * H).expm() # unitary operator for the beamsplitter, time evolution for t=1

# initial state:
psi0 = qt.tensor(qt.fock(n1, 1), qt.fock(n2, 0)) # one photon in mode 1, vacuum in mode 2

# final state:
psi1 = U * psi0
print(psi1)

# setup: two cavities with total photon number = 1. Want swap between two cavities conditioned on whether there is 1 photon in an auxiliary cavity. the auxiliary cavity can be beamsplitter coupled for any duration of time between each of the two other cavities and the beamsplitter can be off for one cavity while it is on for the other cavity. only consider 0 and 1 levels of the auxiliary cavity and 1 exitation manifold of the two main cavities. we also can do arbitrary single qubit rotations on the auxiliary cavity between the 0 and 1 levels.
```
